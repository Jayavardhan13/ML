```python
df = pd.read_csv("/kaggle/input/heights-and-weights-dataset/SOCR-HeightWeight.csv")
```

```python
plt.scatter(df["Weight(Pounds)"],df["Height(Inches)"])
plt.xlabel("Weight(Pounds)")
plt.ylabel("Height(Inches)")
```

```python
x=df[["Weight(Pounds)"]]
y=df["Height(Inches)"]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.80,random_state=40) 
```

```python
# now standardisation 
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(x_train,y_train)
```
